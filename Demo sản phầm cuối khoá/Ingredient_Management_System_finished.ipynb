```python
class Ingredient:
    def __init__(self, name, quantity):
        self.name = name
        self.quantity = quantity
```

```python
from tkinter import *
from tkinter import messagebox

# tạo cửa sổ ứng dụng
window = Tk()

# setup thông tin cho window
window.title("Ingredient Management System")
window.geometry("350x150")

# tạo list để lưu trữ toàn bộ các object
list_ingredient = []

# mô tả các function cho ứng dụng
# add/search/update/delete/reset
def add_ingredient():
    # lấy giá trị từ các widget
    name_value = name.get()
    quantity_value = int(quantity.get())
    
    new_ingredient = Ingredient(name_value, quantity_value)
    
    list_ingredient.append(new_ingredient)
        
    print([item.name for item in list_ingredient])
    
    # hiển thị thông báo
    messagebox.showinfo("Add Ingredient", f"{name_value} is added successfully!")
    
    # xóa giá trị ở các entry
    name.delete(0, END)
    quantity.delete(0, END)
    
def search_ingredient():
    search_name_value = search_name.get()
    
    is_exist = False
    
    for ingredient in list_ingredient:
        if ingredient.name == search_name_value:
            messagebox.showinfo("Search Ingredient", f"Name: {ingredient.name}, quantity: {ingredient.quantity}")
            is_exist = True  # đã tìm thấy nguyên liệu trong hệ thống
            
    if is_exist == False: # không tìm được nguyên liệu nào trong suốt quá trình lặp
        messagebox.showinfo("Search Ingredient", "Ingredient not found!")
    
def update_ingredient():
    name_value = name.get()
    quantity_value = quantity.get()
    
    is_exist = False
    
    for i in range(len(list_ingredient)):
        if list_ingredient[i].name == name_value:
            list_ingredient[i].quantity = quantity_value
            is_exist = True
            messagebox.showinfo("Update Ingredient", f"Update quantity of {name_value} successfully!")
            break
            
    if is_exist == False:
        messagebox.showinfo("Update Ingredient", "Ingredient not found!")
    
def delete_ingredient():
    search_name_value = search_name.get()
    
    is_exist = False
    
    for i in range(len(list_ingredient)):
        if list_ingredient[i].name == search_name_value:
            del list_ingredient[i]
            is_exist = True
            messagebox.showinfo("Delete Ingredient", f"Delete {search_name_value} successfully!")
            break
            
    if is_exist == False:
        messagebox.showinfo("Delete Ingredient", "Ingredient not found!")
    
def reset():
    name.delete(0, END)
    quantity.delete(0, END)
    search_name.delete(0, END)

# setup các widget
# tạo label
name_lbl = Label(window, text="Name:")
quantity_lbl = Label(window, text="Quantity:")

# tạo entry
name = Entry(window)
quantity = Entry(window)
search_name = Entry(window)

# tạo button
add_btn = Button(window, text="Add", command=add_ingredient)
search_btn = Button(window, text="Search", command=search_ingredient)
update_btn = Button(window, text="Update", command=update_ingredient)
delete_btn = Button(window, text="Delete", command=delete_ingredient)
reset_btn = Button(window, text="Reset", command=reset)

# dán các widget lên window
name_lbl.grid(row=0, column=0)
quantity_lbl.grid(row=1, column=0)
name.grid(row=0,column=1)
quantity.grid(row=1,column=1)
add_btn.grid(row=3,column=0)
delete_btn.grid(row=3,column=1)
reset_btn.grid(row=3,column=2)
search_name.grid(row=4,column=0)
search_btn.grid(row=4,column=1)
update_btn.grid(row=4, column=2)

# chạy ứng dụng
window.mainloop()
```
