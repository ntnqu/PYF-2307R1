```python
'''
Quản lý thông tin sinh viên
các thuộc tính như mã sinh viên, họ tên, năm sinh, chuyên ngành, lớp, giới tính

Chức năng:
Thêm sinh viên mới
Cập nhật thông tin của sinh viên
Xóa thông tin sinh viên khỏi
Tìm kiếm thông tin sinh viên
'''
```

```python
class SinhVien:
    def __init__(self, msv, name, year, gender, lop):
        self.msv = msv
        self.name = name
        self.year = year
        self.gender = gender
        self.lop = lop
```

```python
# Bước 1: import thư viện Tkinter
# add module thông báo

from tkinter import *
from tkinter import messagebox

# Bước 2: tạo cửa sổ và setup thông tin cho window
# tạo cửa sổ ứng dụng

window = Tk()
window.title("Hệ Thống Quản Lý Sinh Viên ")
window.geometry('350x180')

# Bước 3: tạo list để lưu trữ toàn bộ các objects trong khi vận hành

data = []

# Bước 4: tạo các hàm xử lý các công việc như add, delete, reset, search, update

# 4.1 : thêm sinh viên

def add_sinh_vien():
    msv_value = msv.get()
    name_value = name.get()
    year_value = int(year.get())
    gender_value = gender.get()
    lop_value = lop.get()
    
    new_sv = SinhVien(msv_value, name_value, year_value, gender_value, lop_value)
    data.append(new_sv)
    

    print([item.name for item in data])
    # print(data)

    #lưu file
    #luu_du_lieu()

    
    # hiện thị thông báo đã add thành công hay chưa
    messagebox.showinfo('Thêm Sinh Viên', f"Họ tên: {name_value}, Mã sinh viên: {msv_value}, Năm sinh: {year_value}, Giới tính: {gender_value}, Lớp: {lop_value} được thêm vào thành công!")


    # xoá các giá trị ở các entry
    name.delete(0, END)
    msv.delete(0, END)
    year.delete(0, END)
    gender.delete(0, END)
    lop.delete(0, END)

# 4.2: hàm xử lý update thông tin sinh viên

def update_sinh_vien():
    msv_value = msv.get()
    name_value = name.get()
    year_value = year.get()
    gender_value = gender.get()
    lop_value = lop.get()
    
    
    is_exist = False
    
    for i in range (len(data)):
        if data[i].msv == msv_value:
            data[i].name = name_value
            data[i].year = year_value
            data[i].gender = gender_value
            data[i].lop = lop_value
            is_exist = True
            messagebox.showinfo("Cập nhật thông tin", f"Cập nhật thông tin của sinh viên {name_value}, mã số {msv_value} thành công!")
            break

    if is_exist == False:
        messagebox.showinfo("Cập nhật thông tin", 'Cập nhật thông tin không thành công!')

    # xoá các giá trị ở các entry
    name.delete(0, END)
    msv.delete(0, END)
    year.delete(0, END)
    gender.delete(0, END)
    lop.delete(0, END)


# 4.3: hàm xử lý tìm kiếm theo msv

def search_sinh_vien():
    search_msv_value = search_msv.get()
    

    is_exist = False
    
    for sinhvien in data:
        if sinhvien.msv == search_msv_value:
            messagebox.showinfo("Tìm Kiếm Thông Tin", f'Họ tên: {sinhvien.name}, Mã sinh viên: {sinhvien.msv}, Năm sinh: {sinhvien.year}, Giới tính: {sinhvien.gender}, Lớp: {sinhvien.lop} được tìm thấy.')
            
            is_exist = True

    if is_exist == False:
        messagebox.showinfo("Tìm Kiếm Thông Tin", f'Mã sinh viên {search_msv_value} không được tìm thấy!')
    
# 4.4: hàm delete theo mã sinh viên

def delete_sinh_vien():
    search_msv_value = search_msv.get()

    is_exist = False
    
    # xoá theo index
    for i in range (len(data)):
        if data[i].msv == search_msv_value:
            del data[i]
            messagebox.showinfo("Xoá Thông Tin", f'Thông tin mã số sinh viên: {search_msv_value} được xoá thành công!')
            is_exist = True
            break

    if is_exist == False:
        messagebox.showinfo("Xoá Thông Tin", 'Không tìm thấy thông tin!')

# 4.5: hàm reset

def reset():
    msv.delete(0, END)
    name.delete(0, END)
    year.delete(0, END)
    gender.delete(0, END)
    lop.delete(0, END)
    search_msv.delete(0, END)

# Bước 5: setup các widget
# tạo các label, có 5 label

Label(window, text = 'Mã Sinh Viên: ').grid(row =0, column =0)
Label(window, text = 'Họ và Tên: ').grid(row =1, column =0)
Label(window, text = 'Năm sinh: ').grid(row =2, column =0)
Label(window, text = 'Giới tính: ').grid(row =3, column =0)
Label(window, text = 'Lớp: ').grid(row =4, column =0)

# tạo các entry, có 6 entry

msv = Entry(window)
name = Entry(window)
year = Entry(window)
gender = Entry(window)
lop = Entry(window)
search_msv = Entry(window)


# tạo các button, có 5 button
add_btn = Button(window, text = 'Thêm', font="timesnewroman 9 bold",command= add_sinh_vien,
                 activeforeground = "green", activebackground = "white").grid(row =5, column =0)
delete_btn = Button(window, text = 'Xoá', font="timesnewroman 9 bold", command= delete_sinh_vien,
                   activeforeground = "green", activebackground = "white").grid(row =5, column =1)
reset_btn = Button(window, text = 'Reset', font="timesnewroman 9 bold", command= reset,
                  activeforeground = "green", activebackground = "white").grid(row =5, column =2)
search_btn = Button(window, text = 'Tìm Kiếm', font="timesnewroman 9 bold", command= search_sinh_vien,
                   activeforeground = "green", activebackground = "white").grid(row =6, column =1)
update_btn = Button(window, text = 'Cập nhật',  font="timesnewroman 9 bold",command= update_sinh_vien,
                   activeforeground = "green", activebackground = "white").grid(row =6, column =2)


# setup vị trí các entry
msv.grid(row =0, column =1)
name.grid(row =1, column =1)
year.grid(row =2, column =1)
gender.grid(row =3, column =1)
lop.grid(row =4, column =1)
search_msv.grid(row =6, column =0)


# Bước 6: chạy ứng dụng
window.mainloop()
```
